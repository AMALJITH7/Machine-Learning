To predict the profit made by a startup on the basis of expenses incurred and the state where they operate

```python
import numpy as np
import pandas as pd
from numpy import math

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
data=pd.read_csv('/content/drive/MyDrive/50_Startups.csv')
data.head()
```

```python
data['New York_state']=np.where(data['State']=='New York',1,0)
data['California_state']=np.where(data['State']=='California',1,0)
data['Florida_state']=np.where(data['State']=='Florida',1,0)
```

```python
data.head()
```

```python
data.drop(['State'],axis=1,inplace=True)
data.head()#The axis parameter is set to 1 to indicate columns; 0 is used for rows.
```

```python
dependent_variable=data['Profit']
dependent_variable.head(5)
```

```python
independent_variable=data[['R&D Spend','Administration','Marketing Spend','New York_state','California_state','Florida_state']]
independent_variable.head(5)
```

```python
y=dependent_variable.values
x=independent_variable.values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train[0:10]
```

```python
y_test[0:10]
```

```python
Scalar=MinMaxScaler()
x_train=Scalar.fit_transform(x_train)
x_test=Scalar.transform(x_test)
```

```python
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
regressor.intercept_
```

```python
y_pred=regressor.predict(x_test)
y_pred
```

```python
y_test
```

```python
MSE=mean_squared_error(y_test,y_pred)
MSE
```

```python
SQRTMSE=math.sqrt(MSE)
SQRTMSE
```

```python
R2=r2_score(y_test,y_pred)
R2
```

```python
print(f'Accuracy is {R2*100}')
```
