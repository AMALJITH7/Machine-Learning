```python
import numpy as np
import pandas as pd
from numpy import math

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error

import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df=pd.read_csv('/content/drive/MyDrive/weight-height.csv')
df
```

```python
df['Male']=np.where(df['Gender']=='Male',1,0)
df['Female']=np.where(df['Gender']=='Female',1,0)
df.head()
```

```python
df.drop(['Gender'],axis=1,inplace=True)
df.head()
```

```python
dependent_variable=df['Height']
dependent_variable.head()
```

```python
independent_variable=df[['Weight','Male','Female']]
independent_variable.head()
```

```python
x=independent_variable.values
y=dependent_variable.values
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train[0:10]
```

```python
y_test[0:10]
```

```python
Scalar=MinMaxScaler()
x_train=Scalar.fit_transform(x_train)
x_test=Scalar.transform(x_test)
```

```python
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
regressor.intercept_
```

```python
regressor.coef_
```

```python
y_pred=regressor.predict(x_test)
y_pred
```

```python
y_test
```

```python
MSE=mean_squared_error(y_test,y_pred)
MSE
```

```python
SQRTMSE=math.sqrt(MSE)
SQRTMSE
```

```python
R2=r2_score(y_test,y_pred)
R2
```

```python
print(f'Accuracy is {R2*100}')
```
